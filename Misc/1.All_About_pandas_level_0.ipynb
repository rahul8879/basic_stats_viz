## Loading and Inspecting Data
We'll start by loading the dataset into a Pandas DataFrame and inspecting its structure.

### Steps:
1. Load the CSV file using `pd.read_csv`.
2. Inspect the first few rows using `head`.
3. Check the overall structure with `info` and `describe`.


```python

# Importing the pandas library
import pandas as pd

# Loading the dataset
file_path = './Data/Salaries.csv'  # Replace with your file path if running locally
data = pd.read_csv(file_path)

type(data)
```

```python
# Inspecting the first few rows
data.tail(3)
```

```python
# Checking the structure of the dataset
data.info()

# Summary statistics for numerical columns
data.describe()
```


## Data Selection and Filtering
To analyze specific aspects of our dataset, we need to know how to select and filter data effectively.

### Steps:
1. Select specific columns or rows using `loc` and `iloc`.
2. Filter rows based on conditions.


```python
# excel ??  basic colum
type(data[['rank', 'salary', 'salary']])
```

```python
data.salary
```

```python
# Selecting specific columns
selected_columns = data[['rank', 'salary']]
selected_columns.head()
```

```python
filtered_df = data[(data.salary > 120000) | (data.discipline == "A")]
filtered_df.head(20)
```

```python
# Filtering rows: Selecting rows where salary is greater than $120,000
high_salary = data[data['salary'] > 120000]
print(high_salary)
```

```python
# Using `loc` for label-based indexing
subset = data.loc[0:4, ['rank', 'salary']]
subset
```

```python
# Using `iloc` for position-based indexing
subset = data.iloc[0:5, 0:3]
subset
```


## Data Manipulation
Pandas allows us to modify our DataFrame by adding, updating, and deleting columns.

### Steps:
1. Add new columns using simple operations.
2. Rename columns for better clarity.
3. Delete unnecessary columns.


```python
data['new-salary']= data['salary']*0.10
data.head(5)
```

```python
# Adding a new column for salary in thousands
data['salary_k'] = data['salary'] / 1000
```

```python
data.rename(columns={'new-salary':'new_salary_updated'},inplace=True)
# test.head(5)
```

```python
# Renaming columns for clarity
data.rename(columns={'phd': 'phd_years', 'service': 'service_years'})
```

```python

# Deleting the newly added column
data.drop('new_salary_updated',axis=1, inplace=True)
```


## Grouping and Aggregation
Grouping data allows us to perform aggregate operations based on specific categories.

### Steps:
1. Group data by a column and calculate aggregates like `mean`, `sum`, etc.
2. Perform multiple aggregate operations.


```python
data['rank'].value_counts()
```

```python
data.groupby(['rank', 'discipline'])['salary'].mean()
```

```python
data.groupby(['rank', 'discipline']).agg({
    'salary' : ['mean','max','min','sum','median']

})
```

```python
# Grouping by rank and calculating average salary
grouped_data = data.groupby('rank')['salary'].mean()
print(grouped_data)
```

```python
# Performing multiple aggregations
agg_data = data.groupby('rank').agg({'salary': ['mean', 'max'], 'phd_years': 'mean'})
print(agg_data)
```


## Visualization
Visualizing data helps uncover patterns and insights. Pandas integrates seamlessly with Matplotlib for basic plots.

### Steps:
1. Plot salary distribution by rank.
2. Create a bar chart for average salary by rank.

```python

import matplotlib.pyplot as plt

# Plotting salary distribution by rank
data.boxplot(column='salary', by='rank', grid=True)
plt.title('Salary Distribution by Rank')
plt.ylabel('Salary')
plt.xlabel('Rank')
plt.show()

```

```python
# Bar chart for average salary by rank
grouped_data.plot(kind='bar', title='Average Salary by Rank')
plt.ylabel('Average Salary')
plt.show()
```
