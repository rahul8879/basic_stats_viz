
# Visualizing Data with Line Graphs and Ogives

In this notebook, we explore two fundamental data visualization techniques: **line graphs** and **ogives**.
We'll use real-world examples to understand the differences between these techniques and their applications.

---

## Why Line Graphs?
Line graphs are used to show trends over time or to depict relationships between data points in a sequential manner.

For example, let's analyze the monthly average high temperatures in Scottsdale, Arizona.



## Raw Data: Monthly Average High Temperatures

Here is the historical average high temperature (in degrees Fahrenheit) for Scottsdale, Arizona:


```python
import pandas as pd
import matplotlib.pyplot as plt


# Raw temperature data

temperature_data = pd.DataFrame({

    "Month": ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"],

    "Average High": [69, 73, 78, 86, 95, 103, 105, 103, 100, 89, 78, 68]

})


# Display the raw data

temperature_data
```


## Line Graph: Average High Temperatures

Let's visualize the monthly average high temperatures using a line graph.


```python

# Line Graph Visualization
plt.figure(figsize=(10, 6))
plt.plot(temperature_data["Month"], temperature_data["Average High"], marker='o', color='#4C72B0', linestyle='-')
plt.title("Monthly Average High Temperatures in Scottsdale, Arizona", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Temperature (F)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```


## What Are Ogives?
An **ogive** is a cumulative line graph that shows the accumulation of data over time. It's especially useful for representing cumulative totals, such as savings or population growth.

Let's explore an example where we track annual investments in a savings account.



## Raw Data: Annual Investments

Here is the investment data where $5,500 is invested annually into a savings account:


```python

# Raw investment data
investment_data = pd.DataFrame({
    "Age": list(range(30, 40)),
    "Annual Investment": [5500] * 10
})

# Calculate cumulative investment
investment_data["Cumulative Investment"] = investment_data["Annual Investment"].cumsum()

# Display the raw data
investment_data
```


## Ogive: Cumulative Investment Over Time

Now, let's visualize the cumulative investment data using an ogive.


```python

# Ogive Visualization
plt.figure(figsize=(10, 6))
plt.plot(investment_data["Age"], investment_data["Cumulative Investment"], marker='o', color='#55A868', linestyle='-')
plt.title("Cumulative Investment Over Time", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Cumulative Savings ($)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
