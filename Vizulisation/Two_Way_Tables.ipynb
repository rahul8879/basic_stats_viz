
# Understanding Two-Way Tables

In this notebook, we explore **two-way tables**, an advanced method for organizing data that depends on two categorical variables. We'll start with one-way tables to build the foundation and then transition to two-way tables to analyze their advantages.

---

## What Are Two-Way Tables?
Two-way tables are used to display data that depends on two independent categories. For example, let's analyze an artist's revenue over four years, broken down by months.




## One-Way Table: Monthly Revenue

Here is the summarized revenue data for an artist, aggregated across four years:


```python
import pandas as pd
# One-Way Table Data
one_way_data = pd.DataFrame({
    "Month": ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"],
    "Revenue": [28361, 14744, 19407, 15891, 21277, 21530, 17990, 21838, 20174, 20025, 48055, 24318]
})

# Display the one-way table
one_way_data
```


## Bar Graph: Monthly Revenue

Let's visualize the one-way table data using a bar graph to identify trends in revenue across months.


```python
import matplotlib.pyplot as plt


# Bar Graph for One-Way Table

plt.figure(figsize=(10, 6))

plt.bar(one_way_data["Month"], one_way_data["Revenue"], color='#4C72B0')

plt.title("Monthly Revenue (One-Way Table)", fontsize=14)

plt.xlabel("Month", fontsize=12)

plt.ylabel("Revenue ($)", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```


## Two-Way Table: Monthly Revenue by Year

Breaking the data down by year provides additional insights. Here's the raw data organized into a two-way table:


```python

# Two-Way Table Data
two_way_data = pd.DataFrame({
    "Month": ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"],
    "2014": [254, 377, 291, 1146, 820, 1138, 694, 486, 538, 448, 2387, 401],
    "2015": [1396, 1084, 1679, 1668, 2708, 3014, 2586, 2080, 3109, 3695, 6495, 2030],
    "2016": [10696, 4745, 7549, 2434, 2326, 2982, 3270, 5451, 6614, 5153, 13128, 7441],
    "2017": [16015, 8538, 9888, 10643, 15423, 14396, 11440, 13821, 9913, 10729, 26045, 14446]
})

# Display the two-way table
two_way_data
```


## Line Graph: Revenue Over Time

Visualizing the revenue data from 2014 to 2017 gives insights into trends and growth over time.


```python
import seaborn as sns
# Prepare Data for Line Graph
two_way_data_long = two_way_data.melt(id_vars=["Month"], var_name="Year", value_name="Revenue")
two_way_data_long["Month_Num"] = pd.Categorical(
    two_way_data_long["Month"],
    categories=["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"],
    ordered=True
).codes + 1

# Line Graph for Two-Way Table
plt.figure(figsize=(12, 6))
sns.lineplot(data=two_way_data_long, x="Month_Num", y="Revenue", hue="Year", marker="o", palette="tab10")
plt.title("Monthly Revenue by Year (Two-Way Table)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.xticks(range(1, 13), ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"])
plt.ylabel("Revenue ($)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
