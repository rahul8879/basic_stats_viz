
# Visualizing Data with Venn Diagrams

In this notebook, we explore **Venn diagrams**, a powerful tool for visualizing overlapping sets of data. Venn diagrams provide a clear way to represent relationships between two or more groups.

---

## Why Venn Diagrams?
Venn diagrams are useful for:
- Representing data with overlapping categories.
- Deriving insights about shared and exclusive characteristics of groups.

### Let's start with a cookies example.



## Cookies Example: Frosting and Sprinkles

We are baking cookies for Christmas and categorizing them as:
- Cookies with frosting.
- Cookies with sprinkles.
- Cookies with both frosting and sprinkles.
- Cookies with neither.

Here is the raw data represented as a two-way table:


```python
import pandas as pd
# Cookies Data: Two-Way Table
cookies_data = pd.DataFrame({
    "Sprinkles": ["Sprinkles", "No Sprinkles", "Total"],
    "Frosting": [4, 5, 9],
    "No Frosting": [3, 4, 7],
    "Total": [7, 9, 16]
}, index=["Frosting", "No Frosting", "Total"])

# Display the data
cookies_data
```


## Venn Diagram: Cookies with Frosting and Sprinkles

The following Venn diagram represents:
- Red circle: Cookies with frosting.
- Gray circle: Cookies with sprinkles.


```python

from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Venn Diagram for Cookies
plt.figure(figsize=(8, 6))
venn = venn2(subsets=(5, 3, 4), set_labels=("Frosting", "Sprinkles"))
plt.title("Venn Diagram: Cookies with Frosting and Sprinkles", fontsize=14)
plt.show()
```


## Pets Example: Dogs, Cats, and Birds

We surveyed students about their favorite pets. Here is the data:
- 2 students like only dogs.
- 3 students like only cats.
- 1 student likes only birds.
- 3 students like dogs and birds, but not cats.
- 5 students like dogs and cats, but not birds.
- 4 students like cats and birds, but not dogs.
- 6 students like all three pets.
- 7 students donâ€™t like any pets.

This data can be represented using a three-circle Venn diagram.


```python

from matplotlib_venn import venn3

# Venn Diagram for Pets
plt.figure(figsize=(10, 8))
venn = venn3(subsets=(2, 3, 5, 1, 3, 4, 6), set_labels=("Dogs", "Cats", "Birds"))
plt.title("Venn Diagram: Students' Favorite Pets", fontsize=14)
plt.show()
```
