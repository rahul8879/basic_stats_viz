# Hypothesis Testing: Supermarket Loyalty Program

This notebook explains the hypothesis testing process step-by-step for the scenario of launching a supermarket loyalty program.

## Problem Statement
A supermarket plans to launch a loyalty program if it results in an average spending per shopper of more than $120 per week.

- A random sample of 80 shoppers enrolled in the pilot program spent an average of $130 per week with a standard deviation of $40.
- The supermarket wants to determine whether the loyalty program should be launched based on this data.


## Step 1: Define Hypotheses
We set up the hypotheses as follows:

- **Null Hypothesis (μ ≤ 120)**: The additional spending per shopper is less than or equal to $120.
- **Alternative Hypothesis (μ > 120)**: The additional spending per shopper is more than $120.

This is a **right-tailed test**, as we are testing whether the spending is significantly greater than $120.

## Step 2: Collect Data
From the problem statement, we have the following:
- Sample size (\(n\)): 80
- Sample mean (\(̄x\)): $130
- Population standard deviation (\(σ\)): $40
- Significance level (\(α\)): 0.05


## Step 3: Compute Test Statistic (t-score)
The formula for the t-score is given as:
\[ t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}} \]
Where:
- \(\bar{x}\): Sample mean
- \(μ_0\): Hypothesized population mean
- \(s\): Standard deviation
- \(n\): Sample size

```python
# Data
sample_mean = 130
hypothesized_mean = 120
std_dev = 40
sample_size = 80

# Calculate t-score
import math
t_score = (sample_mean - hypothesized_mean) / (std_dev / math.sqrt(sample_size))
t_score
```

## Step 4: Determine P-Value
Using the t-distribution, we calculate the p-value for the given t-score. This tells us the probability of obtaining a sample mean as extreme as the one observed, assuming the null hypothesis is true.

```python
from scipy.stats import t

# Calculate p-value (right-tailed test)
df = sample_size - 1  # Degrees of freedom
p_value = 1 - t.cdf(t_score, df)
p_value
```

## Step 5: Make a Decision
Compare the p-value with the significance level (α = 0.05):
- If \(p < α\): Reject the null hypothesis and conclude that the loyalty program increases spending.
- If \(p ≥ α\): Fail to reject the null hypothesis and conclude that there is not enough evidence to support the claim.

```python
# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The loyalty program significantly increases spending.")
else:
    print("Fail to reject the null hypothesis: No significant evidence to support increased spending.")
```

## Step 6: Visualize the Test
We can visualize the t-distribution with the critical value and observed t-score to understand the result.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for visualization
x = np.linspace(-4, 4, 500)
y = t.pdf(x, df)

# Critical value (for alpha = 0.05, right-tailed)
critical_value = t.ppf(1 - alpha, df)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="t-distribution", color="blue")
plt.axvline(critical_value, color="red", linestyle="--", label=f"Critical Value = {critical_value:.2f}")
plt.axvline(t_score, color="green", linestyle="--", label=f"Observed t-score = {t_score:.2f}")
plt.title("Right-Tailed Hypothesis Test")
plt.xlabel("t-value")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# What is a Right-Tailed Hypothesis Test?

A **right-tailed hypothesis test** is used when the alternative hypothesis (\(H_1\)) suggests that the parameter being tested is greater than a specific value. This means the test is looking for evidence that the observed value lies in the **right tail** of the probability distribution.

---

## Key Concepts:

1. **Null Hypothesis (\(H_0\))**:
   - The null hypothesis is the baseline assumption.
   - It generally states that there is no significant effect or difference, or that a parameter is less than or equal to a specified value.
   - For example: \(H_0: \mu \leq \mu_0\), where \(\mu_0\) is the hypothesized mean.

2. **Alternative Hypothesis (\(H_1\))**:
   - The alternative hypothesis represents what we want to prove.
   - In a right-tailed test, \(H_1: \mu > \mu_0\), which indicates that the parameter is significantly greater than the hypothesized value.

3. **Significance Level (\(\alpha\))**:
   - The probability threshold used to decide whether to reject \(H_0\).
   - Commonly set to \(0.05\) (5%).

4. **Right Tail of the Distribution**:
   - The right tail is the area of the distribution to the right of a critical value.
   - If the test statistic falls in this area, it means the observed result is unlikely under \(H_0\), and we reject \(H_0\).

---

## Example of a Right-Tailed Hypothesis Test:

### Scenario:
A supermarket wants to test if their loyalty program increases spending beyond $120 per week.

- **Null Hypothesis (\(H_0\))**: The average spending per customer is \( \mu \leq 120 \) (no significant increase).
- **Alternative Hypothesis (\(H_1\))**: The average spending per customer is \( \mu > 120 \) (significant increase).

If the observed spending (\( \bar{x} \)) is much greater than $120 and the p-value is less than the significance level (\(\alpha = 0.05\)), the null hypothesis \(H_0\) is rejected in favor of \(H_1\).

---

## Right-Tailed Test Visualized:

In a right-tailed test, the **critical region** (rejection region) lies on the **right side** of the distribution:

- If the test statistic (e.g., t-score or z-score) is larger than the critical value, we reject \(H_0\).
- The p-value corresponds to the area under the curve to the right of the test statistic.

---

## Formula for a t-score in a Right-Tailed Test:


\[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
\]

Where:
- \( \bar{x} \): Sample mean
- \( \mu_0 \): Hypothesized mean
- \( s \): Standard deviation
- \( n \): Sample size

---

## Example Decision:

- If \( p < \alpha \): Reject \(H_0\), the program increases spending.
- If \( p \geq \alpha \): Fail to reject \(H_0\), not enough evidence to conclude a significant increase.
