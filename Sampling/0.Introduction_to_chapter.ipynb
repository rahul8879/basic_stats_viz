# Sampling and Inferential Statistics
This notebook introduces the concepts of sampling and inferential statistics with simple and clear examples.

## What is Sampling?
Sampling is the process of selecting a subset (sample) of individuals from a larger population to estimate characteristics of the whole population.

### Why Sampling?
1. Collecting data from the entire population is often impractical or expensive.
2. A well-designed sample can provide accurate insights into the population.

## Example: Sampling Heights in a Town
Let's consider a scenario where you are a researcher interested in estimating the average height of people in a town. Instead of measuring everyone, you decide to collect a random sample and use it to estimate the population mean.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a population
np.random.seed(123)
population_heights = np.random.normal(loc=170, scale=10, size=10000)  # Mean height = 170 cm, Std Dev = 10 cm

# Visualize the population
plt.figure(figsize=(10, 6))
plt.hist(population_heights, bins=50, alpha=0.7, color='blue')
plt.title('Population Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

## Step 2: Draw a Random Sample
We will randomly select 30 individuals from the population and calculate the sample mean.

```python
# Draw a random sample
sample_size = 30
sample = np.random.choice(population_heights, size=sample_size, replace=False)
sample_mean = np.mean(sample)
print(f"Sample Mean: {sample_mean:.2f} cm")
```

## Step 3: Sampling Variation
Each random sample will have a slightly different mean due to variation. Let's simulate this by taking multiple samples and observing their means.

```python
# Sampling variation
num_samples = 1000
sample_means = [np.mean(np.random.choice(population_heights, size=sample_size, replace=False)) for _ in range(num_samples)]

# Visualize the distribution of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, alpha=0.7, color='green')
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean Height (cm)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

## Step 4: Connecting to Inferential Statistics
Inferential statistics allow us to use sample data to make generalizations about the population. The distribution of sample means helps us estimate the population mean with confidence.

### Example: Confidence Interval
A confidence interval provides a range of values within which the true population mean is likely to fall. Let's calculate a 95% confidence interval for the population mean based on a single sample.

```python
from scipy.stats import norm

# Calculate confidence interval
sample_std = np.std(sample, ddof=1)
z_score = norm.ppf(0.975)  # 95% confidence level
margin_of_error = z_score * (sample_std / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(f"95% Confidence Interval: {confidence_interval[0]:.2f} cm to {confidence_interval[1]:.2f} cm")
```

## Conclusion
1. Sampling is a powerful technique for studying populations when collecting data from every individual is impractical.
2. The distribution of sample means approximates a normal distribution, as stated by the Central Limit Theorem.
3. Confidence intervals provide a range within which we expect the population mean to lie, based on the sample data.
