# Understanding the Central Limit Theorem (CLT)
The Central Limit Theorem (CLT) is one of the most important concepts in statistics. It states that the sampling distribution of the sample mean will approach a normal distribution, 
regardless of the original population distribution, as the sample size becomes large enough.

This notebook explores CLT using examples from e-commerce data, such as order amounts, customer ratings, and delivery times.

## Step 1: Simulating E-commerce Data
We will generate synthetic data to simulate an e-commerce scenario, where we have:
- Order amounts (in USD)
- Customer ratings (on a scale of 1 to 5)
- Delivery times (in days)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic e-commerce data
np.random.seed(42)
order_amounts = np.random.exponential(scale=100, size=10000)  # Average order amount
customer_ratings = np.random.uniform(1, 5, size=10000)  # Ratings between 1 and 5
delivery_times = np.random.gamma(shape=2, scale=2, size=10000)  # Delivery times in days

# Create a DataFrame
ecommerce_data = pd.DataFrame({
    'Order Amount': order_amounts,
    'Customer Rating': customer_ratings,
    'Delivery Time': delivery_times
})

ecommerce_data.head()
```

## Step 2: Visualizing the Population Distributions
Let's visualize the distributions of our e-commerce data.

```python
# Visualize the population distributions
plt.figure(figsize=(15, 5))

# Order Amount
plt.subplot(1, 3, 1)
plt.hist(order_amounts, bins=50, alpha=0.7, color='blue')
plt.title('Order Amount Distribution')
plt.xlabel('Order Amount (USD)')
plt.ylabel('Frequency')

# Customer Rating
plt.subplot(1, 3, 2)
plt.hist(customer_ratings, bins=20, alpha=0.7, color='green')
plt.title('Customer Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Delivery Time
plt.subplot(1, 3, 3)
plt.hist(delivery_times, bins=50, alpha=0.7, color='purple')
plt.title('Delivery Time Distribution')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Step 3: Sampling from the Population
We will take multiple random samples from the population and calculate their means to observe the Central Limit Theorem in action.

```python
# Function to generate sampling distribution
def generate_sampling_distribution(data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    return sample_means

# Parameters
sample_size = 50
num_samples = 1000

# Generate sampling distributions
order_amount_means = generate_sampling_distribution(order_amounts, sample_size, num_samples)
rating_means = generate_sampling_distribution(customer_ratings, sample_size, num_samples)
delivery_time_means = generate_sampling_distribution(delivery_times, sample_size, num_samples)
```

## Step 4: Visualizing the Sampling Distributions
Let's visualize the sampling distributions of the means for each variable.

```python
# Visualize the sampling distributions
plt.figure(figsize=(15, 5))

# Order Amount Means
plt.subplot(1, 3, 1)
plt.hist(order_amount_means, bins=30, alpha=0.7, color='blue')
plt.title('Sampling Distribution of Order Amount Means')
plt.xlabel('Mean Order Amount (USD)')
plt.ylabel('Frequency')

# Customer Rating Means
plt.subplot(1, 3, 2)
plt.hist(rating_means, bins=30, alpha=0.7, color='green')
plt.title('Sampling Distribution of Customer Rating Means')
plt.xlabel('Mean Rating')
plt.ylabel('Frequency')

# Delivery Time Means
plt.subplot(1, 3, 3)
plt.hist(delivery_time_means, bins=30, alpha=0.7, color='purple')
plt.title('Sampling Distribution of Delivery Time Means')
plt.xlabel('Mean Delivery Time (Days)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Step 5: Observing the Central Limit Theorem
- Even though the population distributions are not all normal, the sampling distributions of the means are approximately normal.
- This is the Central Limit Theorem in action!

## Conclusion
1. The Central Limit Theorem is a powerful concept that explains why sampling distributions tend to be normal.
