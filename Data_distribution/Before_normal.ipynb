# Understanding PDF and CDF: Explanation with Real-World Example

---

## **1. What are PDF and CDF?**

| **Aspect**         | **PDF**                                 | **CDF**                                     |
|---------------------|-----------------------------------------|---------------------------------------------|
| **Definition**      | Density of the random variable          | Cumulative probability up to \(x\)         |
| **Output**          | Relative likelihood (not a probability) | True probability                           |
| **Range**           | Can take any positive value             | Always between 0 and 1                     |
| **Use Case**        | Describe the shape of the distribution  | Calculate probabilities over ranges        |

---

## **2. Real-World Example: Delivery Times**

### **Scenario**
A delivery service wants to analyze the delivery times for its orders. The delivery times are normally distributed with:
- Mean (\( \mu \)) = 30 minutes,
- Standard deviation (\( \sigma \)) = 5 minutes.

### **Questions**
1. What is the **relative likelihood** of a delivery taking exactly 35 minutes? (Use the PDF.)
2. What is the **probability** that a delivery takes less than 35 minutes? (Use the CDF.)

---

### **Step 1: Formula for PDF**
The formula for the probability density function (PDF) is:
\[
f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
\]

Substitute the values:
- \( x = 35 \),
- \( \mu = 30 \),
- \( \sigma = 5 \).

1. Calculate \( \sigma^2 \):
   \[
   \sigma^2 = 5^2 = 25
   \]

2. Substitute into the formula:
   \[
   f(35) = \frac{1}{\sqrt{2 \pi (25)}} e^{-\frac{(35 - 30)^2}{2 \cdot 25}}
   \]

3. Simplify:
   - \( \sqrt{2 \pi (25)} \approx 12.5331 \),
   - \( (35 - 30)^2 = 25 \),
   - \( 2 \cdot 25 = 50 \),
   - \( e^{-\frac{25}{50}} = e^{-0.5} \approx 0.6065 \).

4. Final Calculation:
   \[
   f(35) = \frac{1}{12.5331} \cdot 0.6065 \approx 0.0484
   \]

### **Answer for PDF**
The relative likelihood of \( X = 35 \) is:
\[
f(35) = 0.0484
\]

---

### **Step 2: Formula for CDF**
The cumulative distribution function (CDF) represents the probability \( P(X \leq x) \). For a normal distribution, we calculate it using the Z-score.

#### Z-Score Formula:
\[
z = \frac{x - \mu}{\sigma}
\]

Substitute the values:
- \( x = 35 \),
- \( \mu = 30 \),
- \( \sigma = 5 \).

\[
z = \frac{35 - 30}{5} = \frac{5}{5} = 1
\]

From the Z-score table:
\[
P(Z \leq 1) = 0.8413
\]

### **Answer for CDF**
The probability that \( X \leq 35 \) is:
\[
F(35) = 0.8413
\]

---

## **3. Interpretation of Results**

1. **PDF**:
   - \( f(35) = 0.0484 \)
   - The relative likelihood of a delivery taking exactly 35 minutes is 0.0484 (not a probability but an indicator of density).

2. **CDF**:
   - \( F(35) = 0.8413 \)
   - There is an **84.13% probability** that a delivery will take **35 minutes or less**.

---

## **4. Key Takeaways**
- **PDF**: Shows the likelihood (density) of a specific value.
- **CDF**: Shows the cumulative probability of a variable being less than or equal to a specific value.
- In the delivery time example:
  - Most deliveries fall within \( 35 \) minutes or less (\( 84.13\% \)).
  - The relative likelihood of exactly \( 35 \) minutes is \( 0.0484 \).
