
# Symmetric and Skewed Distributions, and Outliers

This notebook explores symmetric and skewed distributions, outliers, and their practical applications.
We will cover theoretical aspects along with visualizations and real-world examples, such as analyzing delivery times for Swiggy Instamart.



## Density Curve
A density curve represents the distribution of data. It is the smooth line that encloses the distribution and shows the spread of the data.

For example, we might observe a distribution of ages or delivery times.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Simulate delivery times

np.random.seed(42)

delivery_times = np.random.normal(loc=30, scale=5, size=1000)  # Mean=30 mins, SD=5 mins


# Plot Density Curve

sns.kdeplot(delivery_times, shade=True)
plt.title("Density Curve for Delivery Times")
plt.xlabel("Delivery Time (minutes)")

plt.ylabel("Density")

plt.show()
```


## Symmetric Distributions

Symmetric distributions have equal data distribution on both sides, and the mean, median, and mode are at the center.

Example: On-time deliveries for Swiggy Instamart.


```python

# Generate symmetric data
symmetric_data = np.random.normal(loc=30, scale=5, size=1000)

# Plot Symmetric Distribution
sns.histplot(symmetric_data, kde=True, color='blue')
plt.title("Symmetric Distribution")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")
plt.show()
```


## Skewed Distributions

Skewed distributions are non-symmetric:
- **Left-Skewed (Negative Skew)**: Tail on the left, Mean < Median.
- **Right-Skewed (Positive Skew)**: Tail on the right, Mean > Median.

Example: Delivery delays for Swiggy Instamart during peak hours.


```python

# Right-skewed example (delays)
right_skewed = np.random.exponential(scale=5, size=1000) + 20

# Plot Skewed Distribution
sns.histplot(right_skewed, kde=True, color='green')
plt.title("Right-Skewed Distribution")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")
plt.show()
```


## Outliers

Outliers are data points far from the majority of the data.
We calculate outliers using the **1.5-IQR Rule**:
- Low Outliers: \( Q1 - 1.5 	imes IQR \)
- High Outliers: \( Q3 + 1.5 	imes IQR \)


```python

# Simulated delivery times with outliers
data = pd.Series(np.append(delivery_times, [120, 130]))  # Adding extreme values

# Calculate IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Determine outliers
low_outliers = data[data < Q1 - 1.5 * IQR]
high_outliers = data[data > Q3 + 1.5 * IQR]

print("Low Outliers:", low_outliers)
print("High Outliers:", high_outliers)

# Plot Outliers
sns.boxplot(data)
plt.title("Box Plot of Delivery Times with Outliers")
plt.show()
```


## Summary

Key Learnings:
- Symmetric distributions have equal spread on both sides.
- Skewed distributions show imbalance due to outliers or natural variations.
- Outliers significantly influence measures like mean and standard deviation.
